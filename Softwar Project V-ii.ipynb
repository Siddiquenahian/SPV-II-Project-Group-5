# Matplotlib Example Workflow

```python
# 0. Import matplotlib and get it ready for Ploting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data

x=[1,2,3,4]
y=[12,13,16,17]

# 2. Setup Plot

fig, ax=plt.subplots(figsize=(5,6))

# 3. plot data

ax.plot(x,y)

# 4. Custom data

ax.set(title="Simple Plot",
      xlabel="X-axis",
      ylabel="Y-axis")

# 5. Save and show the whole figure

plt.savefig('Figures\sample-plot.png')
```

# Bar Plot Horizontal

```python
df1=df
fig, ax=plt.subplots(figsize=(7,6))
ax.barh(df1['State'],df1["Total Claim Amount(TK)"]);
ax.set_yticks(df1["State"])
ax.set_xlabel('Total Claim Amount')
ax.set_title('Total Amount By different state',fontweight='bold',fontsize=15)
plt.savefig('Figures\sample-plot.png')
```

```python
df1=df
fig, ax=plt.subplots(figsize=(6,7))
ax.bar(df1['State'],height=df1["Customer Lifetime Value"]);
ax.set_ylabel("Customer Lifetime Value")
ax.set_xlabel('State')
ax.set_title('Show Cuatomer lifetime value By different state',fontweight='bold',fontsize=15)
plt.savefig('Figures\sample-plot.png')
```
